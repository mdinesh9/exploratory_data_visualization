```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
pima_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data')
pima_data.columns = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

## Plotting visuals based on other variables

### Violin Plot

```python
sns.violinplot(x='class',y='age', data=pima_data)
plt.show()
```

### Box Plot

```python
sns.boxplot(x='class',y='age', data=pima_data)
plt.show()
```

### Joint Plot

```python
sns.jointplot(x="plas", y="pres", data=pima_data);
plt.show()
```

### Joint Plot with Regression

```python
sns.jointplot(x="plas", y="pres", data=pima_data, kind = "reg");
plt.show()
```

### Joint plot KDE

```python
sns.jointplot(x="plas", y="pres", data=pima_data, kind="kde");
plt.show()
```

### Pair Plot

```python
sns.pairplot(pima_data);
plt.show()
```

### Three Dimensional

```python
sns.barplot(x="skin", y="age", hue="class", data=pima_data.head(50));
plt.show()
```

### Reg Plot

```python
sns.regplot(x="skin", y="pres", data=pima_data);
plt.show()
```

```python
sns.lmplot(x="skin", y="pres", data=pima_data);
plt.show()
```

```python
sns.lmplot(x="skin", y="pres", data=pima_data, x_estimator=np.mean);
plt.show()
```
