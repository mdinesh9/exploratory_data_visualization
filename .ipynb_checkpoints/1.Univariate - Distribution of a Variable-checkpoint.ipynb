### Plots 

  - Histogram
  - Box and Whisker Plot
  - Density Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

### Read Data

```python
pima_data = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/pima-indians-diabetes/pima-indians-diabetes.data')
pima_data.columns = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

```python
pima_data.head()
```

## Histogram

### Histograms of multiple features

```python
# Increase the figsize and font size if necessary for better visibility
plt.rcParams['figure.figsize'] = (12, 10)
plt.rcParams['font.size'] = 14

pima_data.hist()
plt.show()
```

### Plot histogram of single feature

```python
# Increase the figsize and font size if necessary for better visibility
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 14

pima_data['pres'].hist()
plt.show()
```

## Box and Whisker Plot

### Box and Whisker plot of multiple features

```python
# Increase the figsize and font size if necessary for better visibility
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 14

pima_data.plot(kind='box', subplots=True, layout=(3, 3), sharex=False, sharey=False)
plt.show()
```

### Box and Whisker Plot of single feautre

```python
# Increase the figsize and font size if necessary for better visibility
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 14

pima_data['pres'].plot(kind='box')
plt.show()
```

## Density Plot

### Density Plot of multiple features

```python
# Increase the figsize and font size if necessary for better visibility
plt.rcParams['figure.figsize'] = (15, 10)
plt.rcParams['font.size'] = 14

pima_data.plot(kind='density', subplots=True, layout=(3,3), sharex=False)
plt.show()
```

```python
# Density Plot of single feature
# Increase the figsize and font size if necessary for better visibility
plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 14

pima_data['pres'].plot(kind='density')
plt.show()
```
